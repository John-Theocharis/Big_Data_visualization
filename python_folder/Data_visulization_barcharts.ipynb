```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('/content/augmented_dataset.xlsx')

data
```

```python
data.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the claims per number of jobs
claim_counts = data.groupby('Number of air transportation jobs of state')['Claim Number'].count()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the claim counts per range
claim_counts_per_range = data.groupby(pd.cut(data['Number of air transportation jobs of state'], bins=range(0, 80000, 5000))).size()

# Create the bar chart
plt.bar(claim_counts_per_range.index.astype(str), claim_counts_per_range)

# Set the chart title and axes labels
plt.title('Claim Counts per Range of 5,000 Jobs')
plt.xlabel('Range')
plt.ylabel('Claim Counts')

# Rotate the x-axis labels if needed
plt.xticks(rotation=90)

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the mean of GDP per range
gdp_mean_per_range = data.groupby(pd.cut(data['Value of Current-dollar GDP (thousands of current dollars)'], bins=range(0, 624867909, 50000000)))['GDP'].mean()

# Create the bar chart
plt.bar(gdp_mean_per_range.index.astype(str), gdp_mean_per_range)

# Set the chart title and axes labels
plt.title('Mean GDP per Range of 50,000,000')
plt.xlabel('Range')
plt.ylabel('Mean GDP')

# Rotate the x-axis labels if needed
plt.xticks(rotation=90)

# Display the chart
plt.show()
```
